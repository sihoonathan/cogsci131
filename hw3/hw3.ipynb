###### If you look at Rule #6 & 7, what I basically do is I implemented a function that would be triggered by 
###### either "I love <1>" or "<1> love <2>". When it is triggered, instead of fixing which response correspond 
###### to which rule, I used a random function to choose the response to mimic a more 'free' conversation. 


```python
import re
import random

#Finds if set of specific words are all in the sentence.
def words_finder(words, sentence):
    words_in_setence = re.findall(r'\w+', sentence)
    words_found = []
    for word in words:
        if word in words_in_setence:
            words_found.append(word)
    return len(words) == len(words_found)

#Finds if any word in the list is contained in the sentence.
def word_container(words, sentence):
    words_in_sentence = re.findall(r'\w+', sentence)
    for word in words:
        if word in words_in_sentence:
            return True
    return False

def conversation():
    print(">>> Hi, I'm the Berkeley Campus Guide. Ask me anything you would like to know about UC Berkeley!\n### You can type 'quit' anytime you want to end the chat ### \n ")

    name = ""

    while True:
        user_input = input('>>> ')
        user_input = user_input.lower()
        user_input = user_input.replace("?", "").replace("i'm", "im").replace("!", "")
        words_in_sentence = re.findall(r'\w+', user_input)

        if user_input == 'quit':
            break

        #Rule1 - Anything that includes 'hi' or 'hello' will say the greetings.
        elif re.search(user_input, "hi hello"):
            print("Hi, how are you?")

        #Rule2 - <1> told me that <2> is a good school -> Do you agree with <1> that <2> is a good school?
        elif words_finder(["told", "me"], user_input):
            index = 0
            for word in words_in_sentence:
                if (word == 'me' or 'that') and words_in_sentence[0] == "my":
                    print("Do you agree with your {}".format(words_in_sentence[1]) + " that {}".format(
                        words_in_sentence[index + 5] + " is a good school?"))
                    break
                else:
                    print("Do you agree with them that {}".format(
                        words_in_sentence[index + 4] + " is a good school?"))
                    break

        #Rule3 - <1> <2> a good school?
        elif words_finder(["good", "school"], user_input):

            index = 0
            for word in words_in_sentence:
                if word == 'is' and (words_in_sentence[index + 1] == 'berkeley' or 'cal'):
                    print(words_in_sentence[index + 1].capitalize() + " " + word + " definitely a good school. With notable nobel prize winnders as faculty, we have the most renowned professors in the world. There is a reason why {}".format(words_in_sentence[index+1]) + " is ranked as #1 public university in the world!")
                    break
                index += 1
        #Rule4 - Glad to hear it, <1>
        elif word_container(["good", "fine", "alright", "okay"], user_input):
            if name != "":
                print("Glad to hear it, {}".format(name))
            else:
                print("Glad to hear it!")

        #Rule5 - <1> -> <1> as well.
        elif word_container(["pleasure", "meet", "meeting"], user_input):
            print(user_input.capitalize() + " as well!")

        elif word_container(["im" ], user_input) or words_finder(["i", "am"], user_input) or words_finder(["my", "name"], user_input):
            if name == "":
                name = user_input.strip().split()[-1]
                print("I really like your name, " + user_input.strip().split()[-1].capitalize())
            elif name == user_input.strip().split()[-1]:
                print("I already know you are {}".format(name))
            elif name != user_input.strip().split()[-1]:
                print("Didn't you say your name is {}?".format(name))

        #Rule6 - I love <1> -> Why do you love <1>?
        #Rule7 - <1> love <2> -> Tell me more about <2>
        elif words_finder(["i", "love"], user_input) or word_container(["love", "like"], user_input):
            rand_output = ["Why do you love {}".format(user_input.strip().split()[-1]), "Tell me more about {}".format(user_input.strip().split()[-1])]
            print(random.choice(rand_output))

        #Rule8 - What year was Berkeley <1>? -> Berkeley was <1> in 1868.
        elif word_container(["when", "year", "built", "established"], user_input):
            print("It was {}".format(user_input.strip().split()[-1] + " in 1868"))

        #Rule9 - What is the <1> of Berkeley?
        elif word_container(["mascot"], user_input):
            print("It is Oski")

        #Rule10 - What is the <1> of the school?
        elif word_container(["ranking"], user_input):
            print("Currently, it's ranked 22nd by US News")

        #Rule11 - What is the student to faculty ratio?
        elif words_finder(['ratio', 'faculty'], user_input) or words_finder(['ratio', 'student'], user_input):
            print("It is 18 to 1, which is higher than national average")

        #Rule12 - How <1> <2> are there at Berkeley?
        elif words_finder(['big', 'undergrad'], user_input) or words_finder(['many', 'students'], user_input) or words_finder(['many', 'student'], user_input) or words_finder(['many', 'faculty'], user_input):
            index = 0
            for word in words_in_sentence:
                if word == 'many' and (words_in_sentence[index+1] == 'student' or words_in_sentence[index+1] == 'students'):
                    print("There are approximately 31,000 undergrad and 12,000 grad {}".format(words_in_sentence[index+1]))
                    break
                elif word == 'many' and words_in_sentence[index+1] == 'faculty':
                    print("There are approximately 1,700 {}".format(words_in_sentence[index + 1]) + " members")
                    break
                index += 1

        #Rule13 - How <A> is campus?
        elif word_container(['large'], user_input) or word_container(['big'], user_input):
            for word in words_in_sentence:
                if word == 'large':
                    print("It is as {}".format(word) + " as 1,232 acres")
                    break
                elif word == 'big':
                    print("It is as {}".format(word) + " as 1,232 acres")
                    break

        #Rule14 - <1> is <2> rate? -> <2> rate is ...
        elif word_container(['rate'], user_input):
            index = 0
            for word in words_in_sentence:
                if word == 'acceptance' and (words_in_sentence[index+1] == 'rate' or 'rates'):
                    print(word.capitalize() + " rate to Berkeley is 18%")
                    break
                elif word == 'admission' and (words_in_sentence[index+1] == 'rate' or 'rates'):
                    print(word.capitalize() + " rate to Berkeley is 18%")
                    break
                index += 1

        #Rule15 - <1> <2> <3> <4> cost to go to Berkeley? -> It costs as <2> as ...
        elif words_finder(["much", "cost"], user_input):
            index = 0
            for word in words_in_sentence:
                if word == 'cost' and (words_in_sentence[1] == 'much'):
                    print("It " + word + "s " + "as " + word + " as $13,400 for in-state students and $38,000 for out-of-state students")
                    break
                index += 1

        #Rule16 - <1> <2> majors are there?
        elif words_finder(["how", "many", "majors"], user_input):
            print("There are total of 75 different majors at Berkeley")

        #Rule17 - <1> want to go to <2> -> Why do you want to go to <2>
        elif words_finder(["want", "go"], user_input) or words_finder(["wanna", "go"], user_input):
            print("Why do you want to go to {}?".format(words_in_sentence[-1]))

        #Rule18 - Is <1> anything you <2> <3> me?
        elif words_finder(["anything", "you", "me"], user_input):
            print("Sure! Berkeley is a place where you are surrounded by the most endeavoring people. You get better by being around who are better and smarter than you. Although it could be intimidating sometimes, you can really learn a lot from your peers from interacting with them!")


        #Rule19 - Choice of <1> or <2>
        elif word_container(["yes", "yea", "sure" "nope", "nah", "no"], user_input):
            print(words_in_sentence[0] + "? I respect your answer for what it is")

        #Rule20 - <1> competitive is <2> -> <2> is ...
        elif words_finder(["is", "competitive"], user_input):
            print("{} is extremeley competitive".format(words_in_sentence[-1]))

        #Rule21 - <1> is the <2> <3> at Berkeley? -> <2> <3> at Berkeley is ...
        elif words_finder(["best", "library"], user_input):
            print("Best library at Berkeley is Moffit. It has that modern look I really love!")

        #Rule22 - Is Berkeley really <1> <2> UC? ->
        elif words_finder(["best", "uc"], user_input):
            print("Of course it is THE " + words_in_sentence[-2] + " UC! Don't listen to those UCLA folks saying they are the best!")

        #Rule23 - I have <1> GPA. What's my <2> to get into Berkeley? -> "Your chance is <3> or <4> with <1> GPA"
        elif words_finder(["gpa", "chance"], user_input):
            GPA = re.search(r'[+-]?([0-9]*[.])?[0-9]+', user_input).group()
            if float(GPA) > 3.8:
                print("Your chance is high with {}".format(GPA) + " GPA!")
            else:
                print("Sorry, but your chance is low with {}".format(GPA) + " GPA...")

        #Rule24 - Is <1> better than <2>? -> Berkeley is better than (Choose either <1> or <2> that is not Berkeley)
        elif words_finder(["better", "than"], user_input):
            index = 0
            for word in words_in_sentence:
                if word == 'berkeley' and words_in_sentence[-1] == word:
                    print(word.capitalize() + " is better than {}".format(words_in_sentence[1]))
                    break
                elif word == "berkeley" and words_in_sentence[1] == word:
                    print("Berkeley is better than " + words_in_sentence[-1].capitalize())
                    break

                index += 1

        #25 - Everything Else
        else:
            rand_output = ["roses are red, violets are blue, a face like yours belong in a zoo. don't you worry, i will be there too, not in the cage but laughing at you.", "i have a love-hate relationship with berkeley.", "cogsci 131 is my favorite class!", "i'm not sure what to comment on that..", "did you know that classes at berkeley start 10 minutes after the regular class time? it's called berkeley time. ", "i'm not sure how to answer that", "i'm not sure if i understand your questions"]
            print(random.choice(rand_output))
        print("----------------------------------------------------------------------------------------------------------------------")


conversation()
```
