 
- Exponential function and logarithmic functions are in inverse related. As shown in the above graph, although the value of 10^(-x) would decrease exponentially by the ten-fold for every value of x increased by 1, if you put the y-axis in log scale, it's shown that the graph is decreasing in a straight line. Thus, plotting the curve in log scale for y-axis & looking for a straight line helps to find out if the curve is decreasing or increasing exponentially.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import log

def init_cr(size):
    init_cr = np.zeros([size, 2])
    left_edge, right_edge = np.random.uniform(-10, 0, size), np.random.uniform(10, 0, size)

    for i in range(size):
        init_cr[i, 0], init_cr[i, 1] = left_edge[i], right_edge[i]
    return init_cr

#1
def prob_finder(cr, start=0, val=1):
    one_length, entire_length = [x[1]-x[0] for x in cr if x[0] <= val and x[1] >= val], [x[1]-x[0] for x in cr]
    sum_one_length, sum_entire_length = sum([1/x for x in one_length]), sum([1/x for x in entire_length])

    return sum_one_length / sum_entire_length

#4
def prob_looper(cr, x1, x2):
    x = list(range(x1, x2))
    y = [prob_finder(cr, val=x) for x in range(x1, x2)]
    
    return x, y

init = init_cr(10000)
x, y = prob_looper(init, -5, 6)

plt.plot(x, y, color='orangered', marker = "o")
plt.xlabel('X Values')
plt.ylabel('Probability at value X in Log Scale')
plt.title('Value X Vs. Probability at Value X in Log Scale')
plt.yscale('log')
plt.savefig('foo.png')
plt.show()

x, y = prob_looper(init, -10, 11)
plt.plot(x, y, color='orangered', marker = "o")
plt.xlabel('X Values')
plt.ylabel('Probability at value X in Log Scale')
plt.title('Value X Vs. Probability at Value X in Log Scale')
plt.yscale('log')
plt.savefig('foo.png')
plt.show()
```

- First graphs shows a straight line for x ranging from -5 to 5, which indicates that it has exponential correlation between probability of x in consequential region per every value of x increased by 1. From -5 to 0, it has exponential increase of probability and from 0 to 5, it has exponential decrease of probability. 

- Second graph shows a straight line for x ranging from -10 to 10 and steeper straight line for x ranging from -10 to -5 & 5 to 10, which indicates that this graph not only has an exponential correlation, but also shows that probabilities for x values getting closer to either end, -10 or 10, would be much much less.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import log

def init_cr(size):
    init_cr = np.zeros([size, 2])
    left_edge, right_edge = np.random.uniform(-10, 0, size), np.random.uniform(10, 0, size)

    for i in range(size):
        init_cr[i, 0], init_cr[i, 1] = left_edge[i], right_edge[i]
    return init_cr

#1
def prob_finder(cr, start=0, val=1):
    one_length, entire_length = [x[1]-x[0] for x in cr if x[0] <= val and x[1] >= val], [x[1]-x[0] for x in cr]
    sum_one_length, sum_entire_length = sum([1/x for x in one_length]), sum([1/x for x in entire_length])

    return sum_one_length / sum_entire_length

#2
def prob_looper(cr):
    x = list(range(-10, 11))
    y = [prob_finder(cr, val=x) for x in range(-10, 11)]
    
    return x, y

#5
init_10 = init_cr(10)
x, y = prob_looper(init_10)

plt.plot(x, y, color='orangered', marker = "o")
plt.xlabel('X Values')
plt.ylabel('Probability at value X')
plt.title('10 Consequential Regions')

plt.savefig('foo.png')
plt.show()

init_100 = init_cr(100)
x, y = prob_looper(init_100)

plt.plot(x, y, color='orangered', marker = "o")
plt.xlabel('X Values')
plt.ylabel('Probability at value X')
plt.title('100 Consequential Regions')

plt.savefig('foo.png')
plt.show()

init_1000 = init_cr(1000)
x, y = prob_looper(init_1000)

plt.plot(x, y, color='orangered', marker = "o")
plt.xlabel('X Values')
plt.ylabel('Probability at value X')
plt.title('1000 Consequential Regions')

plt.savefig('foo.png')
plt.show()


```

- The pattern I see is that the more consequential regions there is, the more shape of the left and the right side of the graph mirror each other. This obviously happens because the more consequential regions there is, it will be more likely for the values in the left and right will be averaged out to similar value.  

# 6

- The generalization that I can make is that the higher consequential regions you have, the more accurate your probability graph will be generated. For example, there is definitely a huge difference between 10 regions and 10,000 regions as 10 regions will generate a graph that is skewed and not balanced between left and right side of the graph as opposed to 10,000 regions graph. However, you wouldn't be able to tell the difference between 10,000 and 20,000 because 10,000 is approximate threshold point where probability values do not change as much to make a noticeable difference from graphs with more consequential regions like 20,000 regions.
