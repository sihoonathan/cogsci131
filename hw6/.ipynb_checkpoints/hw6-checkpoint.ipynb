```python
import os
import numpy
import random
import matplotlib.pyplot as plt

# Functions that might be useful (please read the documentation)
# x.flatten() (take a N-dimensional numpy array and make it one-dimensional)
# numpy.random.choice -- choose from the list of images
# numpy.dot -- compute the dot product
# numpy.random.normal -- set up random initial weights

DIM = (28, 28)  # these are the dimensions of the image


def load_image_files(n, path="images/"):
    # helper file to help load the images
    # returns a list of numpy vectors
    images = []
    for f in os.listdir(os.path.join(path, str(n))):  # read files in the path
        p = os.path.join(path, str(n), f)
        if os.path.isfile(p):
            i = numpy.loadtxt(p)
            assert i.shape == DIM  # just check the dimensions here
            # i is loaded as a matrix, but we are going to flatten it into a single vector
            images.append(i.flatten())
    return images


# Load up these image files
A = load_image_files(0)
B = load_image_files(1)

N = len(A[0])  # the total size
assert N == DIM[0] * DIM[1]  # just check our sizes to be sure

# set up some random initial weights
weights = numpy.random.normal(0, 1, size=N)

#Your code here:
NUM_BLOCK = 25

step_func = lambda x: 0 if x < 0 else 1

#This tags image of A with value 0 & B with value 1
training_data = list(zip(A, [0]*len(A))) + list(zip(B, [1]*len(B)))


accuracies = []

def train_func():
    global weights, training_data
    error_list = []

    for x, val in training_data:
        output_sum = numpy.dot(weights, x)
        error = val - step_func(output_sum)
        if error:
            error_list.append(x)
        weights += error * x

    return error_list

def sample_gen(training_data):
    sample_list = []

    for i in range(NUM_BLOCK):
        sample_list.append(random.choice(training_data))

    return sample_list

def cal_accuracy(sample_list):
    errors = 0

    for x, val in sample_list:
        output_sum = numpy.dot(weights, x)
        error = val - step_func(output_sum)
        if error != 0:
            errors += 1

    return 1 - (errors/len(sample_list))

def avg_accuracy():
    average = 0
    for i in range(100):
        average += cal_accuracy(sample_gen(training_data))

    return average/100

accuracy = avg_accuracy()
accuracies.append(accuracy)

i = 0
while True:
    errors = train_func()
    accuracy = avg_accuracy()
    accuracies.append(accuracy)
    i += 1

    if not errors:
        break

average_accuracy = sum(accuracies)/len(accuracies)
print("The average accuracy on blocks of 25 items is: ", average_accuracy)
plt.plot(accuracies,color='orangered', marker = "o")
plt.xlabel("Iterations (n)")
plt.ylabel("Accuracy")
plt.title('Correlation b/t Number of Iterations and Perceptron Accuracy')

plt.savefig("foo.png")
plt.show()

# plt.imshow(numpy.reshape(weights, DIM))
# plt.colorbar()
# plt.savefig("cogscihw13.png")
# plt.show()

# fig = plt.figure()
# ax = fig.add_subplot(111)
# weights.sort()
# ax.plot(weights, range(len(weights)), color="lightblue", linewidth=3, label="example 1 (p=0.25)")
# ax.scatter(weights, range(len(weights)), color="darkgreen", marker="^")
# plt.show()


# plt.plot(x, y, color='orangered', marker = "o")
# plt.xlabel('X Values')
# plt.ylabel('Probability at value X')
# plt.title('Value X Vs. Probability at Value X')

# plt.savefig('foo.png')
# plt.show()
```
